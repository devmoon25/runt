```python
# NORMALIZAR IMAGEN 
img = plt.imread(r"C:\Users\jchinome\OneDrive - Grupo Bancolombia\Auditoria\Automatizacion Querys\Capcha\Limpieza Captcha\img_prueba\MicrosoftTeams-image (11).png")
imagen=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
image = cv2.normalize(imagen, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)
print(image.shape,image.dtype)
plt.imshow(image,cmap="gray")
```

```python
# BINARIZACION O THRESH EN LA IMAGEM 
thresh=cv2.inRange(image,50,255);
plt.imshow(thresh,cmap="gray")
#plt.imshow(lines,cmap="gray")
#cv2.imwrite(r"C:/Users/jchinome/Documents/imagenes1.png",thresh)
```

```python
#MATRIZ DE TRANSFORMACION

kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(2,4)) # MMATRIZ rectagular 
kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT,(2,2)) # MATRIZ cuadrada

  
#DILATACION DE PIXELES BLANCOS
imagenes = cv2.dilate(thresh, kernel,iterations=1) 
imagenes = cv2.dilate(imagenes, kernel1,iterations=1) 


plt.imshow(imagenes,cmap="gray")
cv2.imwrite(r"C:\Users\jchinome\OneDrive - Grupo Bancolombia\Auditoria\Automatizacion Querys\Capcha\Limpieza Captcha\img_prueba\IMG_CLEAN.png",imagenes)
```
